```python
download_path = '/Volumes/workspace/kaggle/kaggle-dataset/kaggle'
dbutils.fs.mkdirs(download_path)
```

```python
df = spark.table('workspace.kaggle.kaggle_1')
creds = df.collect()[0].asDict()
os.environ['KAGGLE_USERNAME'] = creds['username']
os.environ['KAGGLE_KEY'] = creds['key']
```

```python
import kaggle
kaggle.api.dataset_download_files('mkechinov/ecommerce-behavior-data-from-multi-category-store', path=download_path, unzip=True)
```

```python
# Read the CSV files from Volumes (this combines both CSVs automatically)
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/kaggle/kaggle-dataset/kaggle")

# Check what you loaded (optional - yes, this shows both CSV files as one table)
display(df)
print(f"Row count: {df.count()}")

# Write it as a Delta table with overwriteSchema option
df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.bronze.events_raw")

print("âœ… Table created successfully!")
```
